```python
#check for missing values
df.isnull().sum()
```

```python
#replace null values in Review Rating with 
#mean - can be affected by outliers which skews the distribution of the data (not ideal)
#median - robust to outliers (ideal)

#find the median in each category and replace the nulls with that

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#check for null values again
df.isnull().sum()
```

```python
#column names - with upper and lower case and spaces- use snake_casing
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
#check the columns - all columns in snake_case
df.columns
```

```python
#feature engineering :

# 1. create new column - age_group
#this will group customers into 4 age groups(young adult, adult,middle age and senior)
#for analyzing patterns accross different age groups

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
#2. create column purchase_frequency_days
#replace the text form which makes analysis harder when comparing customers or calculate metrics with numerical form making the analysis easier

frequency_mapping = {
    'Weekly': 7,
    'Fortnightly': 14,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Annually': 365
}

df['Purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['Purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
#3. drop the promo_code_used column
df = df.drop('promo_code_used', axis=1)

df.columns
```
